Q3. Check if the price data is skewed.

```python
df['Price'].skew()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Price'], kde=True)
plt.title('Distribution of House Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

### Interpretation of Price Skewness
A positive skewness value (typically above 0.5) and a right-skewed histogram (a long tail on the right side) indicate that the price data is positively skewed. This means there are more houses with lower prices, and a smaller number of houses with very high prices, pulling the average to the right.

Q4. What is the average price for each number of bedrooms?

```python
Avg_price = df.groupby('number of bedrooms')['Price'].mean().sort_values(ascending = False)

print(Avg_price)
```

### Interpretation of Average Price by Number of Bedrooms
This analysis shows the average house price categorized by the number of bedrooms. Typically, houses with more bedrooms are expected to have higher prices, though this trend might not be perfectly linear across all bedroom counts, especially for extreme outliers (e.g., houses with exceptionally many bedrooms that might be unique properties or data anomalies).

Q5.  What is the relationship between living area and price? (Use : Area without Basement).

```python
df['Area of the house(excluding basement)'].corr(df['Price'])
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Area of the house(excluding basement)', y='Price', data=df, alpha=0.6)
plt.title('Relationship between Living Area (excluding basement) and Price')
plt.xlabel('Living Area (excluding basement)')
plt.ylabel('Price')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

### Interpretation of Relationship between Living Area and Price
The correlation coefficient and the scatter plot show the relationship between the living area (excluding basement) and the price of a house. A positive correlation indicates that as the living area increases, the price tends to increase as well. The scatter plot provides a visual representation of this trend, showing the spread and density of data points.

Q9. Which location (zipcode/area) has the highest average price?

```python
highest_avg_price_location = df.groupby('Postal Code')['Price'].mean().sort_values(ascending=False).head(1)
print("Location with the highest average price:",highest_avg_price_location)
```

### Interpretation of Location with Highest Average Price
This analysis identifies the postal code with the highest average house price. This specific location likely represents a premium or highly desirable area, potentially due to factors such as proximity to amenities, good schools, low crime rates, scenic views, or exclusive property types. Understanding these high-value areas is crucial for market analysis and investment decisions.
