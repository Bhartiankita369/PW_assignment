2 : Explain the difference between List and Tuple in Python.

A List in Python is mutable, meaning its elements can be changed, and it is created using [ ]. A Tuple is immutable, meaning its elements cannot be modified, and it is created using ( ). Tuples are faster and used for fixed data, while lists are used when data needs to be updated.
