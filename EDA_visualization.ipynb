```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Titanic.csv')
```

Q1. Create a countplot to show the number of passengers by gender.

```python
plt.figure(figsize=(8,6))
plt.title('number of passengers by gender')
plt.xlabel('gender')
plt.ylabel('count')
sns.countplot(x='Sex',data = df)

plt.show()
```

Q2. Create a bar chart to show the number of passengers in each passenger class using different colors.

```python
plt.title('number of passengers in each passenger class')
plt.xlabel('pclass')
plt.ylabel('count')
pclass_counts = df['Pclass'].value_counts()
plt.bar(x=pclass_counts.index, height=pclass_counts.values)
plt.show()
```

Q3. Create a histogram to show the distribution of passenger age.

```python
sns.histplot(df['Age'])
plt.show()
```

Q4. Create a boxplot to detect outliers in the fare column.

```python
sns.boxplot(y=df['Fare'])
plt.show()
```

Q5. Create a scatterplot to

 a. show the relationship between age and fare.

 b. analyze the relationship between family size (SibSp + Parch) and fare.

```python
# a. show the relationship between age and fare.
sns.scatterplot(x=df['Age'], y=df['Fare'])
plt.show()
```

b. analyze the relationship between family size (SibSp + Parch) and fare.
#Ans

Family Size and Fare Relationship: There appears to be a general trend where passengers traveling with larger family groups tend to pay higher fares. This is likely because more people require more cabin space or multiple tickets, increasing the overall cost. However, there's also a noticeable spread in fares, especially for smaller family sizes (including individuals), indicating other factors like passenger class and cabin location also significantly influence the fare.


```python
# b. analyze the relationship between family size (SibSp + Parch) and fare.
df['FamilySize'] = df['SibSp'] + df['Parch']
sns.scatterplot(x=df['FamilySize'], y=df['Fare'])
plt.title('Relationship between Family Size and Fare')
plt.xlabel('Family Size (SibSp + Parch)')
plt.ylabel('Fare')
plt.show()
```

Q6. Create a countplot to show survival count by gender.

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.show()
```

Q7. Create a bar chart to show survival by passenger class.

```python
plt.title('Survival by passenger class')
plt.xlabel('pclass')
plt.ylabel('survival count')
plt.bar(x=df['Pclass'],height= df['Survived'])
plt.show()
```

Q8. Create a heatmap to show correlation between numerical variables.


```python
plt.title('correlation b/w numerical value')
plt.xlabel('numerical value')
plt.ylabel('numerical variable')
sns.heatmap(df.select_dtypes(include=np.number).corr(),annot=True)
plt.show()
```

Q9. Identify which passenger class paid the highest fare on average.

#Ans
Pclass 1 is paying the highest Fare on average.

```python
df.groupby('Pclass')['Fare'].mean()
```
