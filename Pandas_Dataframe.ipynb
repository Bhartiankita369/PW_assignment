Create a Pandas DataFrame using a Python dictionary with the following columns:
Product,
Region,
Sales,
Quantity.

```python
## 1.Creating a dataframe
import pandas as pd
data = { 'product': ['Headphone','Mobile','Tablet','Laptop','Watch'],
        'region': ['north','south','east','west','north'],
         'Sales': [10000,30000,40000,80000,20000],
         'Quantity':[10,6,5,3,5]

}
df = pd.DataFrame(data)

print(df)
```

Q5. Indexing using .iloc

Using .iloc, display:

1.The first 6 rows and first 4 columns.

2.The Sales value from the 3rd row.

```python
##5. Indexing using .iloc

# 1. First 6 rows and first 4 columns
print(df.iloc[0:6, 0:4])

# 2. Sales value from the 3rd row
print(df.iloc[2, 4])   # row index 2, column index 4 (Sales)
```

Q8. Multiple Condition Filtering

Display rows where:

```python
##8. Multiple Condition Filtering
#Display rows where:Region is West & Sales is greater than 1200

print(df[(df['Region']=='West') & (df['Sales'] > 1200)])
```
