Q7. Build Preprocessing Pipeline

Create a pipeline that:

Applies encoding to categorical columns

Applies scaling to numerical columns

Use:

ColumnTransformer

Pipeline

```python
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler

# Sample dataset (replace with your actual data)
df = pd.read_csv('employee_productivity_dataset.csv')

# Separate features
X = df

# Define column types
categorical_cols = ['Gender', 'Department']
numerical_cols = ['Salary', 'Age']

# Create transformers
categorical_transformer = OneHotEncoder()
numerical_transformer = StandardScaler()

# Combine using ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', categorical_transformer, categorical_cols),
        ('num', numerical_transformer, numerical_cols)
    ]
)

# Create Pipeline
pipeline = Pipeline(steps=[
    ('preprocessing', preprocessor)
])

# Apply pipeline
X_processed = pipeline.fit_transform(X)

# Convert result to DataFrame for display
encoded_cols = pipeline.named_steps['preprocessing'] \
    .named_transformers_['cat'] \
    .get_feature_names_out(categorical_cols)

all_cols = list(encoded_cols) + numerical_cols

df_final = pd.DataFrame(X_processed, columns=all_cols)

print(df_final)
```

Q8. Apply Pipeline

Apply the pipeline on the dataset.

Display:

Transformed dataset

Shape of final dataset

```python
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler

# Sample dataset (replace with your actual dataset)
df = pd.read_csv('employee_productivity_dataset.csv')

# Define columns
categorical_cols = ['Gender', 'Department']
numerical_cols = ['Salary', 'Age']

# Create transformers
categorical_transformer = OneHotEncoder()
numerical_transformer = StandardScaler()

# ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', categorical_transformer, categorical_cols),
        ('num', numerical_transformer, numerical_cols)
    ]
)

# Pipeline
pipeline = Pipeline(steps=[
    ('preprocessing', preprocessor)
])

# Apply pipeline
X_transformed = pipeline.fit_transform(df)

# Convert to DataFrame
encoded_cols = pipeline.named_steps['preprocessing'] \
    .named_transformers_['cat'] \
    .get_feature_names_out(categorical_cols)

final_columns = list(encoded_cols) + numerical_cols

df_final = pd.DataFrame(X_transformed, columns=final_columns)

# Display transformed dataset
print("Transformed Dataset:\n")
print(df_final)

# Display shape
print("\nShape of final dataset:", df_final.shape)
```
