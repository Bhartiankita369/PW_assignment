Q2 Sales Filtering Analysis

Filter the dataset to display:

Orders where Sales > 3500

Orders where Region = "East" and Sales > 2500

Explain what insights you observe.

# Insights
East is a high value market,
because half of the high sales orders coming from East region.

```python
# Orders where Sales > 3500

high_sales = df[df["Sales"] > 3500]
print("order with sales > 3500: ")
print(high_sales)
```

```python
# East region order with sales > 2500

east_high_sales = df[(df["Region"]=='East') & (df['Sales'] >2500)]

print("East region order with sales > 2500: ")
print(east_high_sales)
```

Q7 Pivot Table Creation

Create a pivot table showing:

Rows → Region

Columns → Product

Values → Total Sales

Which region sells the most laptops?

```python
# Create pivot table

pivot_table = pd.pivot_table(
    df,
    values="Sales",
    index="Region",
    columns="Product",
    aggfunc="sum"
)

print(pivot_table)
```
